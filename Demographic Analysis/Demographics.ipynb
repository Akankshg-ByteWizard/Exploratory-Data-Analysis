```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt # matplotlib graph (line,bar,hist,scatter)
import seaborn as sns # (distplot,vilion plot,boxplot,pair plot)
```

```python
conn = sqlite3.connect("WORLD_DATA.db")
cur = conn.cursor()
```

```python
cur.execute('''CREATE TABLE mytable(
   Country               VARCHAR(32) NOT NULL PRIMARY KEY
  ,Number_of_births_2021 INTEGER  NOT NULL
  ,Number_of_deaths_2021 INTEGER  NOT NULL
  ,Suicide_Rate          INTEGER NOT NULL
  ,Fertility_Rate        INTEGER NOT NULL
  ,life_expectancy       INTEGER NOT NULL
);  ''')
```

```python
tab1.to_sql('mytable', conn, if_exists='replace', index = False)
```

```python
cur.execute('select * from mytable')
cur.fetchall()
```

1. Find the Top 10 countries with the highest Number of Birth.

```python
Problem_1= pd.read_sql_query('SELECT Country,Number_of_births_2021 from PROJECT order by Number_of_births_2021 DESC LIMIT 10;',conn)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,8))
plt.bar(Problem_1['Country'], Problem_1['Number_of_births_2021'], edgecolor='red');
plt.xlabel("Country")
plt.ylabel("Number of people Born in 2021")
plt.title("Top 10 Countries accross World in 2021");
```

2. Find the countries and Areas if countries have an area greater than 5,00,000 km.

```python
Problem_2 = pd.read_sql_query('Select Country,Area from Project where area > 500000 order by Area DESC;',conn)
Problem_2
```

```python
plt.figure(figsize=(10,8))
plt.title('Area Occipied By each Country')
plt.xlabel('Area in KM')
sns.barplot(x = Problem_2['Area'],y = Problem_2['Country']);
```

3. Find the 10 least suicide rates in countries with a max fertility rate 
Select TOP 10 .

```python
Problem_3 = pd.read_sql_query('SELECT country,Suicide_Rate,Fertility_Rate from PROJECT order by Fertility_Rate DESC,Suicide_Rate ASC LIMIT 10;',conn)
Problem_3
```

```python
plt.figure(figsize=(10,8))
plt.title('Least Suicide Rate per 100,000 people Where Fertility Rate is Maximum.')
sns.scatterplot(x = Problem_3['Country'],y = Problem_3['Suicide_Rate']);
```

4. Find the Number of birth and country name in each country with minimum Life expectancy.

```python
Problem_4 = pd.read_sql_query("Select Country,Number_of_births_2021,life_expectancy from PROJECT order by life_expectancy;",conn)
Problem_4
```

```python
plt.figure(figsize=(15,30))
plt.xlabel('')
plt.ylabel("Countries")
plt.title("Number of Birth's When Life expectancy was minimum Life-expectancy")

plt.plot(Problem_4['Number_of_births_2021'],Problem_4['Country'],marker='*', color='g', ls='--', lw=3, ms=10, alpha=.9);
```
