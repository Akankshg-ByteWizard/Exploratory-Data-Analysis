## Package Install and Import

### First, we will install and import necessary packages.

```python
#libraries to be installed 
import pandas as pd # for creation of dataframe
import matplotlib.pyplot as plt # matplotlib graph (line,bar,hist,scatter)
import seaborn as sns # (distplot,vilion plot,boxplot,pair plot)

import numpy as np
import matplotlib.patches as mpatches
import warnings
warnings.filterwarnings('ignore')
```

## Loading the dataset

### Now we are ready to load the dataset. We will do this using the standard read_csv command from Pandas. Let's take a glimpse at how the data looks like.

```python
Dataset_1 = pd.read_csv('../601/Dataset_1.csv',header =0)
```

```python
Dataset_2 = pd.read_csv('../601/Dataset_2.csv',header =0)
```

```python
Dataset_2 = Dataset_2.apply(lambda x: x.str.replace(',',''))
Dataset_2 = Dataset_2.apply(lambda x: x.str.replace('%',''))
```

### After a Quick glimpse at the dataset, it looks like a typical World demographic dataset of year 2021

### Here we are loading two datasets i.e Dataset_1 and Dataset_2

```python
FINAL_DATA = Dataset_1.merge(Dataset_2,on = 'Country')
```

```python
FINAL_DATA
```

### After merging both the Dataset_1 and Dataset_2 we get FINAL_DATA.

# Data Preparation and Cleaning


### Dropping all null values as they have no importance in my analysis

```python
FINAL_DATA = FINAL_DATA.dropna()
```

```python
FINAL_DATA.info() # to get the information of the datatype
```

```python
FINAL_DATA = FINAL_DATA.convert_dtypes() # converted the data type , as few columns are having object data type instead of string.
```

```python
FINAL_DATA.info()
```

```python
 # converted the data type , as few columns are having object data type instead of string.
convert_dict = {'area': int,
                'Population':int,
                'Land_area': int
                }
 
FINAL_DATA = FINAL_DATA.astype(convert_dict)
```

```python
FINAL_DATA.info() # to get the infomation of the datatype
```

### There are 167 entries and 12 columns to work with for EDA.

#### We use .describe to get the maximum,minimum and other statistical values in our dataframe.

```python
FINAL_DATA.describe().T
```

```python
FINAL_DATA.nunique()
```

```python
FINAL_DATA
```

```python
FINAL_DATA['Death_Rate'] = (FINAL_DATA['Number of deaths (2021)'] * 100000)/ (FINAL_DATA['Population'] *100)
```

```python
FINAL_DATA.drop(FINAL_DATA[FINAL_DATA['Death_Rate'] < 0].index, inplace = True)
```

```python
FINAL_DATA.reset_index(drop=True, inplace=True)
```

```python
FINAL_DATA
```

# Exploratory Analysis and Visualization

## 1. Highest Number of births (2021) has occured among different countries

#### It'd be interesting to see the comparison between the different countries Number of births in the year of 2021. in this dataset to get an idea of the total Number of births took place among all countries.

```python
x = FINAL_DATA[['Country','Number of births (2021)']] 
x = x.sort_values(by="Number of births (2021)",ascending=False)
```

```python
plt.figure(figsize=(20,35))
plt.title('Comparision b/w Number of births (2021) vs Country')
sns.barplot(x = 'Number of births (2021)',y = 'Country' , data = x);
plt.xlabel("Number of births (2021)", size = 20)
plt.ylabel("Country", size = 20)

for index,data1 in enumerate(x['Number of births (2021)']):
    plt.text(x=data1 , y =index , s=int(f"{data1}") , fontdict=dict(fontsize=10));
```

So there are total 167 Countries in which Majority of highest Number of births  has occured in the year of 2021 is India with 23114000 and minimum Number of births took place in Antigua and barbuda with 1129 births.  

## 2. Comparision graph between LAND Area and AREA

Now, we will explore the Comparision graph between Land Area and Whole world area occupied by countries.

```python
x1 = FINAL_DATA[['Country','Land_area','area']]
x1 = x1.sort_values(by = 'area', ascending  = False)
```

```python
plt.figure(figsize=(12,25))
sns.barplot(x = x1.Land_area , y = x1.Country, alpha=1, palette =['black'])
sns.barplot(x = x1.area, y = x1.Country, alpha = 0.5, palette = ['red'], estimator=sum, ci=None, )
plt.title('Land Area vs Area')
plt.xlabel('Area in kms')
Land_Area = mpatches.Patch(color= 'black', label='Land Area')
area = mpatches.Patch(color='red', label='Area')
plt.legend(handles=[Land_Area,area])
plt.ylabel('Countries');
```

From the above graph it is shown according to the data collected different countries has occupied the Land_area from the area, Comparision is shown.

## 3. Relation between fertility and infant mortality rate.

Here we are finding the relationship between fertility and infant mortatility where if if child mortality is reduced, then eventually fertility reduction will follows.

```python
x2=FINAL_DATA[['Country','Fertility _Rate','Infant_mortality_Rate']]
```

From this graph we can observe that in each country how the fertility rate and infant mortality is icreasing and decreasing.

```python
plt.figure(figsize=(10,30))
ordered_df = x2.sort_values(by='Fertility _Rate')
my_range=range(1,len(x2.index)+1)
sns.set_style("darkgrid")
plt.title('Relationship between fertility and infant mortatility in all coutries')
plt.xlabel('rate per 100,000 people')
plt.ylabel('Countries') 

# The horizontal plot is made using the hline function
plt.hlines(y=my_range, xmin=ordered_df['Fertility _Rate'], xmax=ordered_df['Infant_mortality_Rate'], color='grey', alpha=0.4)
plt.scatter(ordered_df['Fertility _Rate'], my_range, color='skyblue', alpha=1, label='Fertility _Rate')
plt.scatter(ordered_df['Infant_mortality_Rate'], my_range, color='red', alpha=0.4 , label='Infant_mortality_Rate')
plt.yticks(my_range, ordered_df['Country'])
plt.legend(['Fertility _Rate','Infant_mortality_Rate']);
```

From this graph we can observe that in each country how the fertility rate and infant mortality is icreasing and decreasing.

### 4. Relationship between Life Expectancy and Infant Mortality Rate.

Now, we will explore the relationship between Life Expectancy and Infant Mortality Rate.

```python
sns.histplot(FINAL_DATA["life - expectancy"],color="orange", kde=True, linewidth=1)
sns.histplot(FINAL_DATA['Infant_mortality_Rate'], kde=True, linewidth=0)
plt.legend([ "life - expectancy","Infant_mortality_Rate"])
plt.title('Relation between Life-expectancy and Infant_mortality_Rate ')
plt.xlabel('Life-expectancy');
plt.ylabel('Infant_mortality_Rate');
```

From the above graph we can observe that lower values of infant mortality levels lead to higher life expectancy at birth.

## 5. Median age accross the Countries in 2021

We will explore Median age's across all countries 

```python
plt.figure(figsize=(14,7))
g = sns.kdeplot(FINAL_DATA['Median-AGE'], color="Red", shade = True)
g.set_xlabel("age")
g.set_ylabel("Frequency")
plt.title('Median-AGE',size = 20)
plt.show()
```

From the above graph we can observe the average median age is 30.

### 6. Relationship between Life Expectancy and Death Rate.

Now We will take a look at Relationship between Life Expectancy and Death Rate in all countries

```python
x_5 = FINAL_DATA[['Country','life - expectancy','Death_Rate']]
plt.figure(figsize=(10,30))
ordered_df1 = x_5.sort_values(by='Death_Rate')
my_range1=range(1,len(x_5.index)+1)
 
# The horizontal plot is made using the hline function
plt.hlines(y=my_range1, xmin=ordered_df1['Death_Rate'], xmax=ordered_df1['life - expectancy'], color='grey', alpha=0.4)
plt.scatter(ordered_df1['Death_Rate'], my_range1, color='skyblue', alpha=1, label='Death-Rate')
plt.scatter(ordered_df1['life - expectancy'], my_range1, color='green', alpha=0.4 , label='Life-Expectancy')
plt.title('Life Expectancy vs Death Rate',size = 20)
plt.yticks(my_range1, ordered_df1['Country'])
plt.xlabel("rate per 100,000")
plt.ylabel("Countries")
plt.legend(['Death_Rate','life - expectancy']);
```

Life expectancy is related to the average age at death within a population and is inversely related to the population death rates at that time; that is, the lower the death rates the greater the life expectancy.
